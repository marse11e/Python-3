##  Последовательность игры
Чтобы сыграть в Блэкджек, следует выполнить следующие шаги:
1. Создать колоду из 52 карт
2. Перемешать колоду
3. Спросить у Игрока его ставку
4. Убедиться, что ставка Игрока не превышает количество доступных у него фишек
5. Сдать две карты Дилеру и две карты Игроку
6. Показать только одну карту из двух карт Дилера, вторая карта остается скрытой
7. Показать обе карты Игрока
8. Спросить Игрока, хочет ли он взять ещё одну карту (Hit), и если да, то дать ему ещё одну карту
9. Если сумма карт игрока не превышает 21, то спросить его, хочет ли он снова взять ещё одну карту.
10. Если Игрок говорит "хватит" (т.е., дополнительных карт не нужно), то перейти к картам Дилера. Дилер всегда берёт дополнительные карты до тех пор, пока сумма его карт не станет больше или равна 17
11. Определить победителя, и сохранить новое значение для количества фишек у Игрока
12. Спросить Игрока, хочет ли он сыграть снова

**Шаг 2: создайте класс Card**<br>
Объект Card object будет содержать только два атрибута: масть и достоинство. Также можно добавить атрибут значение - "value" - для хранения количества очков. **Шаг 3: Создайте класс Deck**<br>
Здесь мы можем сохранить 52 объекта card в списке, который затем можно перемешать. Но сначала нам нужно *инициализировать* все 52 уникальные карты, и добавить их в наш список. Как только у нас доступен класс Card, мы можем создать объекты Card внутри метода  \_\_init\_\_ для Deck. Сделайте итерации по спискам мастей и достоинств, чтобы создать каждую карту. Это можно сделать внутри класса Deck в методе \_\_init\_\_:

В дополнение к методу \_\_init\_\_, мы можем добавить методы для перемешивания нашей колоды карт, а также для выдачи карт во время игры.<br><br>
ОПЦИОНАЛЬНО: Во время игры Вам не понадобится отображать содержимое колоды карт, однако просмотр карт внутри колоды может пригодиться для отладки программы во время разработки. Для этой цели можно добавить метод \_\_str\_\_ в определение класса.

```python
class Deck:
    
    def __init__(self):
        self.deck = []  # начинаем с пустого списка
        for suit in suits:
            for rank in ranks:
                self.deck.append(Card(suit,rank))  # создаём объекты Card и добавляем их в список
    
    def __str__(self):
        deck_comp = ''  # начинаем с пустой строки
        for card in self.deck:
            deck_comp += '\n '+card.__str__() # добавляем строку print для каждого объекта Card
        return 'В колоде находятся:' + deck_comp

    def shuffle(self):
        random.shuffle(self.deck)
        
    def deal(self):
        single_card = self.deck.pop()
        return single_card
```

```python
class Hand:
    
    def __init__(self):
        self.cards = []  # начинаем с пустого списка, так же, как и в классе Deck 
        self.value = 0   # начинаем со значения 0
        self.aces = 0    # добавляем атрибут, чтобы учитывать тузы
    
    def add_card(self,card):
        self.cards.append(card)
        self.value += values[card.rank]
        if card.rank == 'Туз':
            self.aces += 1  # увеличиваем значение self.aces
    
    def adjust_for_ace(self):
        while self.value > 21 and self.aces:
            self.value -= 10
            self.aces -= 1 
```

Мы добавили в метод add_card увеличение переменной self.aces всякий раз, когда Игроку выпадает Туз. И добавили код в метод adjust_for_aces, который уменьшает количество Тузов каждый раз, когда мы делаем корректировку с 11 на 1, чтобы остаться в пределах 21.

**Шаг 8: Напишите функцию, которая предлагает Игроку взять дополнительную карту или остаться при текущих картах**<br>
Эта функция может принимать колоду и карты игрока hand в качестве параметров, и присваивать значения глобальной переменной playing.<br>
Если Игрок запрашивает дополнительную карту, то вызываем функцию hit(), которую мы обсуждали выше. Если Игрок остается при своих картах, то устанавливаем переменную playing в значение False - с помощью этой переменной мы будем контролировать цикл <code>while</code> позже.

```python
def hit_or_stand(deck,hand):
    global playing  # для контроля цикла while
    
    while True:
        x = input("Вы хотите взять дополнительную карту (h - Hit) или остаться при текущих картах (s - Stand)? Введите 'h' или 's' ")
        
        if x[0].lower() == 'h':
            hit(deck,hand)  # определённая выше функция hit()

        elif x[0].lower() == 's':
            print("Игрок остается при текущих картах. Ход дилера.")
            playing = False

        else:
            print("Извините, пожалуйста попробуйте снова.")
            continue
        break
```

```python
def show_some(player,dealer):
    print("\nКарты Дилера:")
    print(" <карта скрыта>")
    print('',dealer.cards[1])  
    print("\nКарты Игрока:", *player.cards, sep='\n ')
    
def show_all(player,dealer):
    print("\nКарты Дилера:", *dealer.cards, sep='\n ')
    print("Карты Дилера =",dealer.value)
    print("\nКарты Игрока:", *player.cards, sep='\n ')
    print("Карты Игрока =",player.value)
```
