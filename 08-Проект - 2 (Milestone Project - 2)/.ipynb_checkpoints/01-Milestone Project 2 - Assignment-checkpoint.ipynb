# Домашнее задание - Проект 2 - Игра Блэк-Джек (Двадцать одно)
В этом проекте Вы создадите карточную игру BlackJack в Python.

Требования следующие:

* Вы должны создать простую текстовую версию игры [Игра Блэк-Джек (Двадцать одно)](https://ru.wikipedia.org/wiki/Блэкджек)
* В игре будет один игрок против компьютера-дилера (крупье)
* Игрок может остаться при тех же картах (stand) или запросить ещё карту (hit).
* Игрок может выбрать сумму, которую он ставит.
* Нужно следить за общей суммой денег у игрока.
* Вам нужно информировать игрока о выигрышах, проигрышах, переходе через двадцать один и так далее...

И самое главное:

* **Вы должны использовать объектно-ориентированное программирование и классы в части Вашего кода. Нельзя использовать только функции. Используйте классы для колоды карт и для набора карт у игрока. Есть много разных способов сделать это, так что подумайте!**


Также можете дополнить и расширить игру. Например, добавить поддержку нескольких игроков. Реализовать дополнительные правила игры - удвоение ставки (double down), разбиение пар (split) и т.д.! Помните, что Вы можете использовать любые ресурсы для разработки. И, как всегда:
